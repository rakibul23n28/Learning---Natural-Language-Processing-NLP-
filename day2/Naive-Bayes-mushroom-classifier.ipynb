#posterior probability
##likelihood
##prior probability

```python
#prior probability
def prior_prob(y_train,label):
    count = np.sum(y_train == label)
    total = len(y_train)
    return count/total
```

```python
#likelihood

def cond_prob(X_train, y_train, feature_col, feature_val, label):
    x_filtered = X_train[y_train == label]
    numerator = np.sum(x_filtered[:, feature_col] == feature_val)
    denominator = x_filtered.shape[0]
    return float(numerator / denominator)
```

```python
def predict(X_train, y_train,X_test):
    classes = np.unique(y_train)
    n_features = X_train.shape[1]
    posterior_prob = []
    for label in classes:
        likelihood = 1.0
        for feature in range(n_features):
            likelihood *= cond_prob(X_train, y_train, feature, X_test[feature], label)
        prior = prior_prob(y_train, label)
        posterior = likelihood * prior
        posterior_prob.append(posterior)
    pred = np.argmax(posterior_prob)
    return pred
```

```python
def accuracy(X_test, y_test, X_train, y_train):
    pred = []
    for i in range(len(X_test)):
        pred.append(predict(X_train, y_train, X_test[i]))
    
    y_pred = np.array(pred)
    acc = np.sum(y_pred == y_test) / len(y_test)
    return acc
```

```python
acc = accuracy(X_test, y_test, X_train, y_train)
print(f"Accuracy: {acc * 100:.4f}%")
```
