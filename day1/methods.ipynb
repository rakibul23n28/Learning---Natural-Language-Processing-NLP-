```python
from sklearn.datasets import fetch_20newsgroups
```

```python
text_data = fetch_20newsgroups()
```

```python
import numpy as np
```

```python
raw_text = text_data.data[:4]
raw_text
```

#stage 1 convart into lower text

```python
clear_text_1 = []

def to_lower_case(text):
    for text in raw_text:
        clear_text_1.append(str.lower(text))
to_lower_case(raw_text)
clear_text_1[:4]
```

#stage 2 tokenize

```python

from nltk.tokenize import sent_tokenize , word_tokenize

import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
sent_tok = []
for text in clear_text_1:
    sent = sent_tokenize(text)
    sent_tok.append(sent)
```

#word tokenize

```python
clear_text_2 = [ word_tokenize(text) for text in clear_text_1]
clear_text_2
```

```python
import re

clear_text_3 = []

for words in clear_text_2:
    clean = []
    for word in words:
        if re.match(r'^[a-zA-Z]+$', word):
            clean.append(word)
    clear_text_3.append(clean)

clear_text_3
```

#stage 4 stopwords removal

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
clean_text_4 = []

for words in clear_text_3:
    clean = []
    for word in words:
        if word not in stopwords.words('english'):
            clean.append(word)
    clean_text_4.append(clean)

clean_text_4
```

#stage 5 stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
port  = PorterStemmer()
```

```python
a = [port.stem(word) for word in ['playing', 'played', 'plays', 'player','washing', 'washed', 'washes', 'washer','running', 'ran', 'runs', 'runner']]
a
```

```python
clean_text_5 = []

for words in clean_text_4:
    clean = []
    for word in words:
        clean.append(port.stem(word))
    clean_text_5.append(clean)

clean_text_5
```

#Lemitization

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
wnet = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
lem = []
for words in clean_text_4:
    clean = []
    for word in words:
        clean.append(wnet.lemmatize(word))
    lem.append(clean)
lem
```
